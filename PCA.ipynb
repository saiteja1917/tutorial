```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = load_iris()
features = iris.data
target = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Standardize the features
scaler = StandardScaler()
features_standardized = scaler.fit_transform(features)

# Apply PCA
pca = PCA(n_components=2)  # Use 2 components for visualization
principal_components = pca.fit_transform(features_standardized)

# Create a DataFrame with PCA results and target species
pc_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
pc_df['species'] = target_names[target]

# Plot PCA results
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pc_df, x='PC1', y='PC2', hue='species', palette='viridis', s=100)
plt.title('PCA of Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Species')
plt.show()

# Additional scatter plot of Sepal Length vs Sepal Width (original data)
plt.figure(figsize=(10, 6))
df_original = pd.DataFrame(data=features, columns=feature_names)
df_original['species'] = target_names[target]
sns.scatterplot(data=df_original, x='sepal length (cm)', y='sepal width (cm)', hue='species', palette='viridis', s=100)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```
